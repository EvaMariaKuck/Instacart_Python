## 1 Importing data files and libraries

```python
# Importing libraries

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Importing data file

path = r'C:\Users\Lenovo\Documents\Instacart Basket Analysis'
df_ords_prods_merge = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merged_4.csv'))
```

## 2 Creating charts

```python
# Create a bar chart

df_ords_prods_merge['orders_day_of_week'].value_counts().plot.bar()
```

```python
# Create a bar chart with sorted indeces

df_ords_prods_merge['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
# Create a bar chart with colors

df_ords_prods_merge['orders_day_of_week'].value_counts().plot.bar(color=['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

```python
# Saving the bar chart in Jupyter's memory

bar = df_ords_prods_merge['orders_day_of_week'].value_counts().plot.bar(color =['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

```python
# Saving the bar chart as an image file

bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

```python
# Histogram

df_ords_prods_merge['prices'].plot.hist(bins = 25)
```

```python
# Scatterplot

sns.scatterplot(x = 'prices', y = 'prices',data = df_ords_prods_merge)
```

```python
# Further checks

df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 100]
```

```python
# Turning the outliers into missing values

df_ords_prods_merge.loc[df_ords_prods_merge['prices'] >100, 'prices'] = np.nan
```

```python
# Checking the maximum value

df_ords_prods_merge['prices'].max()
```

```python
# Redoing the histogram

df_ords_prods_merge['prices'].plot.hist(bins = 25)
```

```python
# A histogram with more granularity

df_ords_prods_merge['prices'].plot.hist(bins = 70)
```

```python
# Saving the histogram as an image file

hist = df_ords_prods_merge['prices'].plot.hist(bins = 70)
hist.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_prices.png'))
```

```python
# Create a list holding True/False values to the test np.random.rand <= 0.7

np.random.seed(4)
dev = np.random.rand(len(df_ords_prods_merge)) <= 0.7
```
