### Question 2: Creating the variable "busiest days"

```python
# Creating an empty list refering to the frequencies of "orders_day_of_week" from above: 
# Saturday (=0) and Sunday (=1) are the busiest days, Wednesday (=4) and Tuesday (=3) are the slowest days. 
# Every other day is regularly busy.

result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest days")
  elif value == 1:
    result.append("Busiest days")
  elif value == 4:
    result.append("Slowest days")
  elif value == 3:
    result.append("Slowest days")
  else:
    result.append("Regularly busy")
```

```python
# Creating the new variable

df_ords_prods_merged['busiest_day_2'] = result
```

### Question 4: Create a new column called “busiest_period_of_day” containing labels “Most orders,” “Average orders,” and “Fewest orders”

```python
# Checking the frequencies

df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Creating an empty list refering to the frequencies of "order_hour_of_day" from above:
# The top 8 hours of the day (in terms of highest frequencies of orders) get the label "Most orders", 
# the bottom 8 hours of the day get the label "Fewest orders" 
# and the 8 hours of the day in between the two get the label "Average orders".

result = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value == 3:
    result.append("Fewest orders")
  elif value == 4:
    result.append("Fewest orders")  
  elif value == 2:
    result.append("Fewest orders")  
  elif value == 5:
    result.append("Fewest orders")  
  elif value == 1:
    result.append("Fewest orders")
  elif value == 0:
    result.append("Fewest orders")
  elif value == 6:
    result.append("Fewest orders")
  elif value == 23:
    result.append("Fewest orders")
  elif value == 9:
    result.append("Most orders")
  elif value == 16:
    result.append("Most orders")
  elif value == 12:
    result.append("Most orders")
  elif value == 13:
    result.append("Most orders")
  elif value == 15:
    result.append("Most orders")
  elif value == 14:
    result.append("Most orders")
  elif value == 11:
    result.append("Most orders")
  elif value == 10:
    result.append("Most orders")
  else:
    result.append("Average orders")
```

```python
# Creating the new variable

df_ords_prods_merged['busiest_hour_of_day'] = result
```
